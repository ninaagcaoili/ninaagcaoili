```python

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}
```

```python
#Problem 3: Routing 
#return est time to go from place 1 to place 2 
#key is in a tuple, index 1 is the place and 2 is the destination
#circular route (one way)

def eta(legs,source,destination):
    
    time_estimate=0

    while source != destination:
        for place, time in legs.items():
            if place[0]==source: 
                time_estimate+=time["travel_time_mins"]
                source=place[1]
                break
                
    return time_estimate    
            

eta(legs,"upd","dlsu")
```
